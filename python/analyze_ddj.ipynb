# Evaluate Updated Text

```python
from collections import Counter

# Define the file path
file_path = r"..\public\docs\ddj_guodian_chu.txt"  # Adjust if needed

# Read the file content
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()

# Count character occurrences
char_counts = Counter(text)

# Sort by frequency in descending order
sorted_chars = sorted(char_counts.items(), key=lambda x: x[1], reverse=True)

# Display results
for char, count in sorted_chars:
    print(f"'{char}': {count}")
```

```python
print(len(sorted_chars))
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Set the font family to a font that supports Chinese characters
rcParams['font.family'] = 'SimSun'
rcParams['axes.unicode_minus'] = False  # Ensure minus sign is displayed correctly

meaningful_list = '道德大常不下知生死天地自圣王人民无有为'

# Extract characters and counts
characters = [char for char, count in sorted_chars if char in meaningful_list]
counts = [count for char, count in sorted_chars if char in meaningful_list]

# Plot the histogram
plt.figure(figsize=(12, 6), dpi=300)
plt.bar(characters, counts, color='#0F766E')
plt.title('Character Frequency', fontsize=16, fontfamily='Georgia')
# plt.xlabel('Characters', fontsize=14, fontfamily='Georgia')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Number of Occurrences', fontsize=14, fontfamily='Georgia')
plt.gca().set_xlim(left=-0.5, right=len(meaningful_list) - 0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.show()
```

# Compare Updated Text vs Standard Text

```python
# Compare character counts across the two texts
sorted_chars_compare = []

characters = [char for char, count in sorted_chars if char in meaningful_list]
for char in meaningful_list:
    for pair in sorted_chars:
        if pair[0] == char:
            cnt_updated = pair[1]
            break

    for pair in sorted_chars_std:
        if pair[0] == char:
            cnt_std = pair[1]
            break

    diff = cnt_updated - cnt_std

    # Append the character and the difference in counts
    sorted_chars_compare.append((char, cnt_updated, cnt_std, diff))

# Sort by frequency in descending order
sorted_chars_compare.sort(key=lambda x: x[1], reverse=True)

sorted_chars_compare
```

```python
import numpy as np

# Extract data from sorted_chars_compare
chars = [x[0] for x in sorted_chars_compare]
cnt_updated = np.array([x[1] for x in sorted_chars_compare])
cnt_std = np.array([x[2] for x in sorted_chars_compare])

# Create figure and axis
plt.figure(figsize=(12, 6), dpi=300)

# Calculate the minimum values for overlap
overlap = np.minimum(cnt_updated, cnt_std)
diff_positive = np.maximum(cnt_updated - cnt_std, 0)  # Green areas
diff_negative = np.maximum(cnt_std - cnt_updated, 0)  # Red areas

# Create the stacked bar chart
plt.bar(chars, overlap, color='teal', label='Overlapping Characters')
plt.bar(chars, diff_negative, bottom=overlap, color='#d3696d', label='Additional Characters in Standard Text')
plt.bar(chars, diff_positive, bottom=overlap, color='#31c848', label='Additional Characters in Mawangdui Text')
plt.title('Character Frequency', fontsize=16, fontfamily='Georgia')
# plt.xlabel('Characters', fontsize=14, fontfamily='Georgia')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Number of Occurrences', fontsize=14, fontfamily='Georgia')
plt.gca().set_xlim(left=-0.5, right=len(meaningful_list) - 0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.legend(prop={'family': 'Georgia', 'size': 14})
# Adjust layout to prevent label cutoff
plt.tight_layout()
plt.show()
```
